```python
file_index_list = [
    {
        "path": "/Users/zhixuan/Desktop/temp/Data2_排放量.xlsx",
        "focused_index": "CO2 Emissions (kg)",
        "columns": ["CO2 Emissions (kilotons)", "NOX Emissions (tons)", "SO2 Emissions (tons)", "CO2 Emissions (kg)"]
    },
    {
        "path": "/Users/zhixuan/Desktop/temp/Data1_发电量.xlsx",
        "focused_index": "Electricity Generation (TWh)",
        "columns": ["Heat Generation (TJ)", "Electricity Generation (GWh)", "Planned Curtailment (GWh)",
                    "Electricity Generation (TWh)"]
    }
]

ref = RefData("/Users/zhixuan/Desktop/temp/color_index_simplified.xlsx")
data = WorkingData(file_index_list, ref)
data.rule_out('省份', selected_tech_group=['压缩空气储能', '抽水蓄能', '化学储能'], tech_group_select_reverse=True)
# data.draw("Electricity Generation (GWh)", focused_region='Heilongjiang', scatter=True)

df = data.working_df.groupby(by=['年份', '省份']).sum().reset_index()
```

```python
df
```

```python
years = list(set(df['年份']))
for year in years:
    co2 = df[(df['年份']==year) & (df['省份']=='East-InnerMongolia')]['CO2 Emissions (kg)']
    gen = df[(df['年份']==year) & (df['省份']=='East-InnerMongolia')]['Electricity Generation (TWh)']
    mongolia = {
        '年份': year,
        '省份': 'InnerMongolia',
        'CO2 Emissions (kg)': co2,
        'Electricity Generation (TWh)': gen
    }
    df = pd.concat([df, pd.DataFrame(mongolia)], ignore_index=True)
```

```python
df['Complex'] = df['CO2 Emissions (kg)'] / df['Electricity Generation (TWh)']
df = df.drop(columns=['CO2 Emissions (kg)', 'Electricity Generation (TWh)'])
```

```python
df = df.pivot(index='省份', columns='年份', values='Complex')
```
